# Install matplotlib and import pyplot as plt 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# plot a basic line chart using plt.plot()

```python
x =[1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'b+--')
plt.show()
```

# add title and labels to axes

```python
x =[1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]

plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='Green',linewidth=5,linestyle='dotted')
plt.show()
```

# change line color,style and markers

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
plt.show()
```

# plot multiple lines in the same chart

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="average")
plt.show()
```

# add a legend to the chart 

```python
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="average")
plt.legend()
plt.show()
```

# save plot as an image file 

```python
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="average")
plt.legend()
plt.savefig('pic1.PNG',transparent=None,facecolor='auto',edgecolor='auto')
plt.show()
```

# plot bar char using plt.bar()

```python
import numpy as np
categories = np.array(["Category A","Category B","Category C","Category D"])
values = np.array([10,15,20,25])
plt.bar(categories,values)
plt.xlabel("categories")
plt.ylabel("values")
plt.title("simple bar chart")
plt.show()
```

# plot a histogram using plt.hist()

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.hist(days,bins=[80,100,125,150],rwidth=0.1,color='g')
plt.show()
```

# plot a pie chart using plt.pie() 

```python
x=[1,2,3,4,5,6,7]
y=[8,9,3,5,6,8,9]
xpos=np.arange(len(x))
xpos
plt.bar(xpos-0.2,x,width=0.4,label='X')
plt.bar(xpos+0.2,y,width=0.4,label='Y')
plt.xticks(xpos,x)
plt.ylabel('y(Bin)')
plt.title('Task')
plt.legend()
plt.show()
```

# use subplots to display multiple plots 

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("Plot1")

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("Plot2")
plt.show()
```

# customize ticks and grade lines

```python
import numpy as np
x=np.arange(0,10,1)
y=np.sin(x)
plt.plot(x,y)
plt.xticks([0,2,4,6,8])
plt.yticks([-1,0,1])
plt.title('sine wave')
plt.show()
```

# use different figure size

```python
plt.figure(figsize=(10,6))
plt.plot([1,2,3],[4,5,6])
plt.title("Custom figure size example")
plt.show()
```

# plot a scatter plot using plt.scatter()

```python
import numpy as np
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)
plt.show()
```

# add annotations to plot points

```python
x = [1,2,3,4]
y = [2,5,3,6]
plt.scatter(x,y)
for i,j in zip(x,y):
    plt.annotate(f'({i},{j})',(i,j))
plt.show()
```

# create a stacked bar chart

```python
x=['A','B','C','D']
y1 = [10,20,10,30]
y2 = [20,25,15,25]
plt.bar(x,y1,color='r')
plt.bar(x,y2,bottom=y1,color='b')
plt.show()
```

# change background and axis colors

```python
fig = plt.figure(facecolor='lightblue')
plt.plot([1,2,3],[4,5,6])
plt.show()
```

# create a plots with twin axes(twinx,twiny)

```python
x = [1,2,3]
y1 = [10,20,30]
y2 = [100,200,300]
fig, ax1 = plt.subplots()
ax1.plot(x,y1,'b-')
ax1.set_ylabel('y1',color='b')

ax2 = ax1.twinx()
ax2.plot(x,y2,'r-')
ax2.set_ylabel('y2',color='r')
plt.show()
```

# use different styles(e.g., 'ggplot','seaborn')

```python
plt.style.use('ggplot')

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)
plt.title('GGPlot Style')
plt.show()
```

```python
import seaborn as sns
sns.set()
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.title('Seaborn style')
plt.show()
```
